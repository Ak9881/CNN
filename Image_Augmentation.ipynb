```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
img = mpimg.imread('Tiger.jpg')
```

```python
plt.imshow(img)
```

# Rescaling

```python
img_scale = resize(img, (150, 150))
```

```python
plt.imshow(img_scale)
```

# Flipping

```python
horizontal_flip = np.fliplr(img_scale)
```

```python
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img_scale)
```

```python
plt.imshow(vertical_flip)
```

```python
from skimage import transform
trans_img = transform.rotate(img, random.uniform(-40,40))
plt.imshow(trans_img)
```

# Adding Noise

```python
img_noise = random_noise(img, mode='s&p', clip = True)
plt.imshow(img_noise)
```
